```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_bits = 256
downgrade_level = 2

def update_pixel_value(pixel):
    pixel = round((pixel/(image_bits-1)) * (round(image_bits/downgrade_level) -1))

    return pixel

update_values = np.vectorize(update_pixel_value)
    
img = cv2.imread("ctskull-256.tif", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap="gray", vmin=0, vmax=255)



```

```python
img_128 = update_values(img)
plt.imshow(img_128, cmap="gray", vmin=0, vmax=127)
image_bits = image_bits/downgrade_level
```

```python
img_64 = update_values(img_128)
plt.imshow(img_64, cmap="gray", vmin=0, vmax=63)
image_bits = image_bits/downgrade_level
```

```python
img_32 = update_values(img_64)
plt.imshow(img_32, cmap="gray", vmin=0, vmax=31)
image_bits = image_bits/downgrade_level
```

```python
img_16 = update_values(img_32)
plt.imshow(img_16, cmap="gray", vmin=0, vmax=15)
image_bits = image_bits/downgrade_level
```

```python

img_8 = update_values(img_16)
plt.imshow(img_8, cmap="gray", vmin=0, vmax=7)
image_bits = image_bits/downgrade_level
```

```python


img_4 = update_values(img_8)
plt.imshow(img_4, cmap="gray", vmin=0, vmax=3)
image_bits = image_bits/downgrade_level
```

```python

img_2 = update_values(img_4)
plt.imshow(img_2, cmap="gray", vmin=0, vmax=1)
```
